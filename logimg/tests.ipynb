```python
from skimage import data
from matplotlib import pyplot as plt
import numpy as np
```

```python
camera=data.camera()
moon=data.moon()
camera=np.array(data.camera().tolist())
moon=np.array(data.moon().tolist())

plus=camera+moon

for i in range(len(plus)):
    for j in range(len(plus[0])):
        if plus[i][j]>255:
            plus[i][j]=255
plt.imshow(camera,cmap='gray',interpolation='nearest')
plt.show()
plt.imshow(moon, cmap='gray', interpolation='nearest') 
plt.show()
plt.imshow(plus, cmap='gray', interpolation='nearest') 
plt.show()
```

```python
camera=data.camera()
moon=data.moon()
suma=LIPSpace().sum(moon,camera)
plt.imshow(camera,cmap='gray',interpolation='nearest')
plt.show()
plt.imshow(moon, cmap='gray', interpolation='nearest') 
plt.show()
plt.imshow(suma, cmap='gray', interpolation='nearest') 
plt.show()
```

```python
import numpy as np

camera=data.camera()

t_camera=np.fft.fft2(camera)
i_t_camera=np.abs(np.fft.ifft2(t_camera))

plt.imshow(camera,cmap='gray',interpolation='nearest')
plt.show()
plt.imshow(i_t_camera, cmap='gray', interpolation='nearest') 
plt.show()
```

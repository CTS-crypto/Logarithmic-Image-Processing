```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from skimage.util import random_noise
from skimage.restoration import denoise_tv_chambolle

a=io.imread('000000038118.jpg',as_gray=True)
b=io.imread('CXR7_IM-2263-1001.png',as_gray=True)

plt.imshow(a,cmap='gray',interpolation='nearest')
plt.show()
plt.imshow(b, cmap='gray', interpolation='nearest') 
plt.show()
```

```python
from models.lip import LIPImage,LIPSpace
from models.hlip import HLIPImage,HLIPSpace
from models.pslip import PSLIPImage,PSLIPSpace
from models.plip import PLIPSpace
from models.ppslip import PPSLIPSpace
```

## Gaussian Noise

```python
nga=random_noise(a,mode='gaussian',var=0.01)
ngb=random_noise(b,mode='gaussian',var=0.01)

io.imsave('nga.png',nga)
io.imsave('ngb.png',ngb)

plt.imshow(nga,cmap='gray',interpolation='nearest')
plt.show()

plt.imshow(ngb, cmap='gray', interpolation='nearest')
plt.show()
```

## Poisson Noise

```python
npa=random_noise(a,mode='poisson',seed=42)
npb=random_noise(b,mode='poisson',seed=42)

io.imsave('npa.png',npa)
io.imsave('npb.png',npb)

plt.imshow(npa,cmap='gray',interpolation='nearest')
plt.show()

plt.imshow(npb, cmap='gray', interpolation='nearest')
plt.show()
```

## Lineal

### TV

```python
tldnga=denoise_tv_chambolle(nga,weight=0.1)

io.imsave('tldnga.png',tldnga)

plt.imshow(tldnga,cmap='gray',interpolation='nearest')
plt.show()

tldngb=denoise_tv_chambolle(ngb,weight=0.1)

io.imsave('tldngb.png',tldngb)

plt.imshow(tldngb, cmap='gray', interpolation='nearest')
plt.show()

tldnpa=denoise_tv_chambolle(npa,weight=0.1)

io.imsave('tldnpa.png',tldnpa)

plt.imshow(tldnpa,cmap='gray',interpolation='nearest')
plt.show()

tldnpb=denoise_tv_chambolle(npb,weight=0.1)

io.imsave('tldnpb.png',tldnpb)

plt.imshow(tldnpb, cmap='gray', interpolation='nearest')
plt.show()
```

## LIP

### TV

```python
js=LIPSpace(1)

jnga=js.function(js.gray_tone(nga))

tjdnga=js.inverse_gray_tone(js.inverse_function(denoise_tv_chambolle(jnga,weight=0.1)))

io.imsave('tjdnga.png',tjdnga)

plt.imshow(tjdnga,cmap='gray',interpolation='nearest')
plt.show()

jngb=js.function(js.gray_tone(ngb))

tjdngb=js.inverse_gray_tone(js.inverse_function(denoise_tv_chambolle(jngb,weight=0.1)))

io.imsave('tjdngb.png',tjdngb)

plt.imshow(tjdngb, cmap='gray', interpolation='nearest')
plt.show()

jnpa=js.function(js.gray_tone(npa))

tjdnpa=js.inverse_gray_tone(js.inverse_function(denoise_tv_chambolle(jnpa,weight=0.1)))

io.imsave('tjdnpa.png',tjdnpa)

plt.imshow(tjdnpa,cmap='gray',interpolation='nearest')
plt.show()

jnpb=js.function(js.gray_tone(npb))

tjdnpb=js.inverse_gray_tone(js.inverse_function(denoise_tv_chambolle(jnpb,weight=0.1)))

io.imsave('tjdnpb.png',tjdnpb)

plt.imshow(tjdnpb, cmap='gray', interpolation='nearest')
plt.show()
```

## HLIP

### TV

```python
hs=HLIPSpace(1.0000001)

hnga=hs.function(hs.gray_tone(nga))

thdnga=hs.inverse_gray_tone(hs.inverse_function(denoise_tv_chambolle(hnga,weight=0.1)))

io.imsave('thdnga.png',thdnga)

plt.imshow(thdnga,cmap='gray',interpolation='nearest')
plt.show()

hngb=hs.function(hs.gray_tone(ngb))

thdngb=hs.inverse_gray_tone(hs.inverse_function(denoise_tv_chambolle(hngb,weight=0.1)))

io.imsave('thdngb.png',thdngb)

plt.imshow(thdngb, cmap='gray', interpolation='nearest')
plt.show()

hnpa=hs.function(hs.gray_tone(npa))

thdnpa=hs.inverse_gray_tone(hs.inverse_function(denoise_tv_chambolle(hnpa,weight=0.1)))

io.imsave('thdnpa.png',thdnpa)

plt.imshow(thdnpa,cmap='gray',interpolation='nearest')
plt.show()

hnpb=hs.function(hs.gray_tone(npb))

thdnpb=hs.inverse_gray_tone(hs.inverse_function(denoise_tv_chambolle(hnpb,weight=0.1)))

io.imsave('thdnpb.png',thdnpb)

plt.imshow(thdnpb, cmap='gray', interpolation='nearest')
plt.show()
```

## Metrics

### PSNR

```python
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage import img_as_float

fa=img_as_float(a)
fb=img_as_float(b)

print(psnr(nga,fa))
print(psnr(tldnga,fa))
print(psnr(tjdnga,fa))
print(psnr(thdnga,fa))
print(psnr(tpsdnga,fa))
```

### SSIM

```python
from skimage.metrics import structural_similarity as ssim

print(ssim(nga,fa))
print(ssim(tldnga,fa))
print(ssim(tjdnga,fa))
print(ssim(thdnga,fa))
print(ssim(tpsdnga,fa))
```

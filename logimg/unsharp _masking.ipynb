## PLIP

```python
mus=[256,400,500,800,1000,1026,4100,600,300]
lambd=256

ps=PLIPSpace()

gamma=256

image_list=[]

pa=ps.function(ps.gray_tone(a),lambd)

aux=ps.inverse_function(scharr(pa),lambd)

image_list.append(ps.gray_tone(a))
image_list.append(aux)

d_1 = 4 if image_list[0].shape[0]%4==0 else 5
d_2 = 4 if image_list[0].shape[1]%4==0 else 5
max_emee=-1
max_mu=-1

result_sum=ps.sum(image_list[0],image_list[1],gamma)
for mu in mus:
    max_pi_r=np.max(result_sum)
    if mu < max_pi_r:
        continue
    result=ps.inverse_gray_tone(result_sum,mu)
    actual_emee=emee(result,1,d_1,d_2,0.5)
    if actual_emee > max_emee:
        max_emee=actual_emee
        max_mu=mu
pa_spa=ps.inverse_gray_tone(ps.sum(image_list[0],image_list[1],gamma),max_mu)
print(max_emee)
print(max_mu)

io.imsave('pa_spa.png',pa_spa)
pa_spa=io.imread('pa_spa.png')
print(contrast_img(pa_spa,256,2))
plt.imshow(pa_spa,cmap='gray',interpolation='nearest')
```

```python
from skimage import data
import matplotlib.pyplot as plt
from skimage import io

a=io.imread('000000037751.jpg')
b=io.imread('000000038118.jpg')
c=io.imread('000000074457.jpg')
d=io.imread('000000074860.jpg')

plt.imshow(a,cmap='gray',interpolation='nearest')
plt.show()
plt.imshow(b, cmap='gray', interpolation='nearest') 
plt.show()
plt.imshow(c, cmap='gray', interpolation='nearest') 
plt.show()
plt.imshow(d, cmap='gray', interpolation='nearest') 
plt.show()
```

```python
from models.lip import LIPImage,LIPSpace
from models.hlip import HLIPImage,HLIPSpace
from models.patrascu import PatrascuImage,PatrascuSpace
from models.pslip import PSLIPImage,PSLIPSpace
from models.plip import PLIPSpace
from skimage.metrics import peak_signal_noise_ratio,structural_similarity
```

## LIP

```python
js=LIPSpace()

jsa_3=js.inverse_gray_tone(js.s_mul(js.gray_tone(a),3))
jsa_05=js.inverse_gray_tone(js.s_mul(js.gray_tone(a),0.5))

plt.imshow(jsa_3,cmap='gray',interpolation='nearest')
plt.show()
plt.imshow(jsa_05, cmap='gray', interpolation='nearest') 
plt.show()
```

### PSNR

```python
print(peak_signal_noise_ratio(jsa_3,a,data_range=255))
print(peak_signal_noise_ratio(jsa_05,a,data_range=255))
```

### SSIM

```python
print(structural_similarity(jsa_3,a,data_range=255))
print(structural_similarity(jsa_05,a,data_range=255))
```

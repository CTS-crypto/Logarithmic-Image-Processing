## Imagen original

```python
from skimage import data
import matplotlib.pyplot as plt

camera=data.camera()
plt.imshow(camera, cmap='gray', interpolation='nearest') 
plt.show()
```

```python
plt.hist(camera)
plt.show()
```

## Jourlin-Pinoli Model

```python
from lip import LIPImage

lip_camera=LIPImage(camera)
lip_camera.show()
```

```python
lip_camera.histogram()
```

## SLIP Model

```python
from slip import SLIPImage

slip_camera=SLIPImage(camera)
slip_camera.show()
```

```python
slip_camera.histogram()
```

## Patrascu Model

```python
from patrascu import PatrascuImage

patrascu_image=PatrascuImage(camera)
patrascu_image.show()
```

```python
patrascu_image.histogram()
```

## Imagen original

```python
from skimage import data
import matplotlib.pyplot as plt
from skimage import io

a=io.imread('000000037751.jpg')
b=io.imread('000000038118.jpg')
c=io.imread('000000074457.jpg')
d=io.imread('000000074860.jpg')

plt.imshow(a,cmap='gray',interpolation='nearest')
plt.show()
plt.imshow(b, cmap='gray', interpolation='nearest') 
plt.show()
plt.imshow(c, cmap='gray', interpolation='nearest') 
plt.show()
plt.imshow(d, cmap='gray', interpolation='nearest') 
plt.show()
```

```python
from models.lip import LIPImage,LIPSpace
from models.hlip import HLIPImage,HLIPSpace
from models.patrascu import PatrascuImage,PatrascuSpace
from models.pslip import PSLIPImage,PSLIPSpace
```

## Jourlin-Pinoli Model

```python
ja=LIPImage(a)
jb=LIPImage(b)
jc=LIPImage(c)
jd=LIPImage(d)

ja_jb=(ja+jb).transform()
jc_jd=(jc+jd).transform()

print(ja_jb)
print(jc_jd)

plt.imshow(ja_jb,cmap='gray',interpolation='nearest')
plt.show()
plt.imshow(jc_jd, cmap='gray', interpolation='nearest') 
plt.show()
```

```python
js=LIPSpace()

jsab=js.inverse_gray_tone(js.sum(js.gray_tone(a),js.gray_tone(b)))
jscd=js.inverse_gray_tone(js.sum(js.gray_tone(c),js.gray_tone(d)))

print(jsab)
print(jscd)

plt.imshow(jsab,cmap='gray',interpolation='nearest')
plt.show()
plt.imshow(jscd, cmap='gray', interpolation='nearest') 
plt.show()
```

## HLIP Model

```python
ha=HLIPImage(a)
hb=HLIPImage(b)
hc=HLIPImage(c)
hd=HLIPImage(d)

ha_hb=(ha+hb).transform()
hc_hd=(hc+hd).transform()

print(ha_hb)
print(hc_hd)

plt.imshow(ha_hb,cmap='gray',interpolation='nearest')
plt.show()
plt.imshow(hc_hd, cmap='gray', interpolation='nearest') 
plt.show()
```

```python
from skimage import data
import matplotlib.pyplot as plt
from skimage import io

a=io.imread('000000037751.jpg')
b=io.imread('000000038118.jpg')
c=io.imread('000000074457.jpg')
d=io.imread('000000074860.jpg')

plt.imshow(a,cmap='gray',interpolation='nearest')
plt.show()
plt.imshow(b, cmap='gray', interpolation='nearest') 
plt.show()
plt.imshow(c, cmap='gray', interpolation='nearest') 
plt.show()
plt.imshow(d, cmap='gray', interpolation='nearest') 
plt.show()
```

```python
from models.lip import LIPImage,LIPSpace
from models.hlip import HLIPImage,HLIPSpace
from models.patrascu import PatrascuImage,PatrascuSpace
from models.pslip import PSLIPImage,PSLIPSpace
from models.plip import PLIPSpace
from skimage.metrics import peak_signal_noise_ratio,structural_similarity
```

## Lineal

```python
import numpy as np

sab=np.array(a.tolist())+np.array(b.tolist())
scd=np.array(c.tolist())+np.array(d.tolist())

if np.max(sab)>255:
    sab=sab*255/np.max(sab)

if np.max(scd)>255:
    scd=scd*255/np.max(scd)

plt.imshow(sab,cmap='gray',interpolation='nearest')
plt.show()
plt.imshow(scd, cmap='gray', interpolation='nearest') 
plt.show()
```

## LIP

```python
js=LIPSpace()

jsab=js.inverse_gray_tone(js.sum(js.gray_tone(a),js.gray_tone(b)))
jscd=js.inverse_gray_tone(js.sum(js.gray_tone(c),js.gray_tone(d)))

plt.imshow(jsab,cmap='gray',interpolation='nearest')
plt.show()
plt.imshow(jscd, cmap='gray', interpolation='nearest') 
plt.show()
```

### PSNR

```python
print(peak_signal_noise_ratio(jsab,sab,data_range=255))
print(peak_signal_noise_ratio(jscd,scd,data_range=255))
```

### SSIM

```python
print(structural_similarity(jsab,a,data_range=255))
print(structural_similarity(jsab,b,data_range=255))

print('-------------------------------------------')

print(structural_similarity(sab,a,data_range=255))
print(structural_similarity(sab,b,data_range=255))

print('-------------------------------------------')

print(structural_similarity(jsab,sab,data_range=255))

print('-------------------------------------------')
print('-------------------------------------------')

print(structural_similarity(jscd,c,data_range=255))
print(structural_similarity(jscd,d,data_range=255))

print('-------------------------------------------')

print(structural_similarity(scd,c,data_range=255))
print(structural_similarity(scd,d,data_range=255))

print('-------------------------------------------')

print(structural_similarity(jscd,scd,data_range=255))
```
